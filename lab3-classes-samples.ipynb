```python

# Ex. 1 Sample solution:
class CloseApproach:
    """
    Represent a close approach of an object to an orbiting body.

    Attributes:
        orbiting_body (str): The name of the body being approached (e.g., Earth, Mars).
        approach_date (datetime): The date and time of the close approach.
        miss_distance (float): The miss distance in kilometers.
        relative_velocity (float): The relative velocity in
            kilometers per hour.
    """
    def __init__(self, orbiting_body: str, approach_date: datetime,
                 miss_distance: float, relative_velocity: float):
        self.orbiting_body = orbiting_body
        self.approach_date = approach_date
        self.miss_distance = miss_distance
        self.relative_velocity = relative_velocity
        
    def __str__(self):
        return f"""The object approached or will approach {self.orbiting_body}
On: {self.approach_date}
Miss distance: {self.miss_distance} km
Relative velocity: {self.relative_velocity} km/h
"""
```

```python
# Ex. 2 Sample solution:
def get_close_approach(record):
    """
    Create a CloseApproach object from a record dictionary.

    Args:
        record (dict): A dictionary containing close approach data.

    Returns:
        CloseApproach: A CloseApproach instance created from the record.
    """
    return CloseApproach(record['orbiting_body'],
                         datetime.datetime.fromisoformat(record['close_approach_date']).date(),
                         float(record['miss_distance']["kilometers"]),
                         float(record['relative_velocity']['kilometers_per_hour']))

print(get_close_approach(neo['close_approach_data'][0])) # Print to test
```
